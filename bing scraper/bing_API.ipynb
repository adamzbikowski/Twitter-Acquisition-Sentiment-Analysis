```python
subscription_key = "60d6f85b673641dc929c99424abac994"
search_url = "https://api.bing.microsoft.com/v7.0/news/search"
```

```python
search_term = "Elon Musk"
```

```python
headers = {"Ocp-Apim-Subscription-Key" : subscription_key}
params  = {"q": search_term, 'count': 100, "textDecorations": True, "textFormat": "HTML"}
```

```python
response = requests.get(search_url, headers=headers, params=params)
response.raise_for_status()
search_results= response.json()
```

```python
print(search_results)
```

```python
descriptions = [article["description"] for article in search_results["value"]]
```

```python
html_table = "<table border='1'><tr><th>Position</th><th>Title</th><th>Source</th><th>Date</th><th>Description</th><th>URL</th></tr>"

position = 1  # Initialize the position variable

for article in search_results["value"]:
    title = article["name"] if "name" in article else ""
    source = article["provider"][0]["name"] if "provider" in article else ""
    formatted_dates = datetime.strptime(article["datePublished"], '%Y-%m-%dT%H:%M:%S.0000000Z').strftime('%Y-%m-%d')
    descriptions = article["description"] if "description" in article else ""
    url = article["url"] if "url" in article else ""

    html_table += f"<tr><td>{position}</td><td>{title}</td><td>{source}</td><td>{formatted_dates}</td><td>{descriptions}</td><td><a href='{url}' target='_blank'>{url}</a></td></tr>"

    position += 1  # Increment the position for each article

html_table += "</table>"

# Display the HTML table
display(HTML(html_table))
```
